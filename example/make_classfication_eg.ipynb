```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X, y = make_classification(n_samples=300, n_features=2, n_redundant=0, n_informative=2, random_state=25,
                           n_clusters_per_class=1, scale=100)
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

```python
from sklearn import svm
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
```

```python
X = preprocessing.scale(X)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=1/3,random_state=7)
```

```python
svm_classifier = svm.SVC()
svm_classifier.fit(X_train,y_train)
svm_classifier.score(X_test,y_test)
```
